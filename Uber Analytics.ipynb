```python
import mysql.connector
```

```python
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password= 'system',
    database="uber_project"
)
```

```python
cursor = connection.cursor()
```

```python
# Report on the average trips / rider and the distribution of riders with different TpR buckets
query = "SELECT client_uuid, AVG(trip_distance_miles) AS avg_trips_per_rider FROM uber_tripdata GROUP BY client_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
# Report on the average trips / rider and the distribution of riders with different TpR buckets for subscription trips
query = "SELECT client_uuid, AVG(trip_distance_miles) AS avg_trips_per_rider FROM uber_tripdata WHERE is_subs_trip = 'TRUE' GROUP BY client_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
# Report on the fulfillment rates (% of trips that got completed) and the cancellation rates by rider and by driver
query = "SELECT client_uuid, AVG(avg_trips_per_rider) AS avg_trips_per_rider_subscription FROM (SELECT client_uuid, AVG(trip_distance_miles) AS avg_trips_per_rider FROM uber_tripdata WHERE is_subs_trip = 'TRUE' GROUP BY client_uuid) AS tpr_subquery;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
# Report on the distribution of fulfillment rate by hour of day and by day of week
query = "SELECT client_uuid, COUNT(CASE WHEN status = 'completed' THEN 1 ELSE NULL END) AS completed_trips, COUNT(*) AS total_trips, (COUNT(CASE WHEN status = 'completed' THEN 1 ELSE NULL END) / COUNT(*)) * 100 AS fulfillment_rate FROM uber_tripdata GROUP BY client_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
# Report on the distribution in number of trips over the duration between the request and the trip for both micro-scheduling and pre-scheduling requests
query = "SELECT client_uuid, (SUM(CASE WHEN status = 'completed' THEN 1 ELSE 0 END) / SUM(1)) * 100 AS fulfillment_rate FROM uber_tripdata GROUP BY client_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python


query = "SELECT client_uuid, COUNT(*) AS total_trips FROM uber_tripdata GROUP BY client_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
query = "SELECT client_uuid, COUNT(CASE WHEN status = 'rider_canceled' THEN 1 ELSE NULL END) AS cancelled_by_rider, COUNT(*) AS total_trips, (COUNT(CASE WHEN status = 'rider_canceled' THEN 1 ELSE NULL END) / COUNT(*)) * 100 AS cancellation_rate_by_rider FROM uber_tripdata GROUP BY client_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
query = "SELECT driver_uuid, COUNT(CASE WHEN status = 'driver_canceled' THEN 1 ELSE NULL END) AS cancelled_by_driver, COUNT(*) AS total_trips,(COUNT(CASE WHEN status = 'driver_canceled' THEN 1 ELSE NULL END) / COUNT(*)) * 100 AS cancellation_rate_by_driver FROM uber_tripdata WHERE driver_uuid !='' GROUP BY driver_uuid;"
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
query = '''SELECT HOUR(begintrip_date_time) AS hour_of_day,
       DAYOFWEEK(begintrip_date_time) AS day_of_week,
       COUNT(CASE WHEN status = 'completed' THEN 1 ELSE NULL END) AS completed_trips,
       COUNT(*) AS total_trips,
       (COUNT(CASE WHEN status = 'completed' THEN 1 ELSE NULL END) / COUNT(*)) * 100 AS fulfillment_rate
FROM uber_tripdata WHERE begintrip_date_time !=''
GROUP BY HOUR(begintrip_date_time), DAYOFWEEK(begintrip_date_time);'''
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
query = '''SELECT TIMESTAMPDIFF(MINUTE, request_date_time, begintrip_date_time) AS trip_duration, request_type FROM uber_tripdata WHERE request_type IN ('microscheduling', 'prescheduling');'''
cursor.execute(query)
result = cursor.fetchall()
print(result)
```

```python
query = ''SELECT vehicle_capacity, AVG(trips_per_capacity) AS avg_utilization
FROM (
        SELECT vehicle_capacity, COUNT(*) / vehicle_capacity AS trips_per_capacity
        FROM uber_tripdata WHERE vehicle_capacity !=''
        GROUP BY vehicle_uuid
    ) AS trips_per_capacity_subquery
    GROUP BY vehicle_capacity;''
cursor.execute(query)
result = cursor.fetchall()
print(result)
```
